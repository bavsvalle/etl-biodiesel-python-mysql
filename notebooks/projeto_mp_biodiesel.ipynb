```python
import pandas as pd
# Importação pacote pandas
```

```python
df_mp_original = pd.read_csv('biodiesel-materia-prima.csv')
# Leitura arquivo csv
```

```python
df_mp_original.head()
# Etapa 1: Limpeza
# Verificando as primeiras 5 linhas do dataframe
```

```python
df_mp_original.info()
# Verificando as info do dataframe
```

```python
df_mp_original.columns
# Verificando quais as colunas do dataframe
```

```python
df_mp_original.shape
# Verificando quant. de linhas e colunas
```

```python
df_mp_original.isnull().sum()
# Verificando quantos dados vazios em cada coluna
```

```python
df_mp_original.columns = df_mp_original.columns.str.lower()
df_mp_original.head()

# Padronizando str em minúscula
```

```python
df_mp_original['região'] = df_mp_original['região'].str.lower()
df_mp_original['produto'] = df_mp_original['produto'].str.lower()
df_mp_original['estado'] = df_mp_original['estado'].str.lower()
df_mp_original.head()

# Padronizando str em minúscula
```

```python
!pip install unidecode
from unidecode import unidecode
df_mp_original = df_mp_original.map(lambda x: unidecode(x) if isinstance(x, str) else x)
df_mp_original.head()

# Removendo acentuação linhas
```

```python
df_mp_original.columns = [unidecode(str(col)) for col in df_mp_original.columns]
df_mp_original.head()

# Removendo acentuação colunas
```

```python
df_mp_original.rename(columns={"quantidade (m3)": "quantidade_m3"}, inplace=True)

df_mp_original.head()
```

```python
df_mp_original['mes'] = pd.to_datetime(df_mp_original['mes/ano'], format='%m/%Y').dt.month
df_mp_original['ano'] = pd.to_datetime(df_mp_original['mes/ano'], format='%m/%Y').dt.year
df_mp_original.head()

# Separando mês e ano em duas colunas
```

```python
df_ano = pd.DataFrame(df_mp_original['ano'].unique(), columns=['ano'])
df_ano.head()

# Criando dataframe de anos únicos
```

```python
df_ano['ano_id'] = df_ano.index + 1
df_ano.head()

# Incluindo coluna de id de cada ano único
```

```python
df_mp_original = df_mp_original.merge(df_ano, on=['ano'], how='left')
df_mp_original.head()

# Incluindo coluna de id de cada ano
```

```python
df_mes = pd.DataFrame(df_mp_original['mes'].unique(), columns=['mes'])
df_mes.head()

# Criando dataframe de meses únicos
```

```python
df_mes['mes_id'] = df_mes.index + 1
df_mes.head()

# Incluindo coluna de id de cada mês único
```

```python
df_mp_original = df_mp_original.merge(df_mes, on=['mes'], how='left')
df_mp_original.head()

# Incluindo coluna de id de cada mês
```

```python
df_regioes_unicas = pd.DataFrame(df_mp_original['regiao'].unique(), columns=['regiao'])
df_regioes_unicas.head()

# Criando dataframe de regiões únicas
```

```python
df_regioes_unicas['regiao_id'] = df_regioes_unicas.index + 1
df_regioes_unicas.head()

# Incluindo coluna de id de cada região única
```

```python
df_mp_original = df_mp_original.merge(df_regioes_unicas, on=['regiao'], how='left')
df_mp_original.head()

# Incluindo coluna de id de cada região
```

```python
df_estados_unicos = pd.DataFrame(df_mp_original['estado'].unique(), columns=['estado'])
df_estados_unicos.head()

# Criando dataframe de estados únicos
```

```python
df_estados_unicos['estado_id'] = df_estados_unicos.index + 1
df_estados_unicos.head()

# Incluindo coluna de id de cada estado único
```

```python
df_mp_original = df_mp_original.merge(df_estados_unicos, on=['estado'], how='left')
df_mp_original.head()

# Incluindo coluna de id de cada estado
```

```python
df_produtos_unicos = pd.DataFrame(df_mp_original['produto'].unique(), columns=['produto'])
df_produtos_unicos.head()

# Criando dataframe de produtos únicos
```

```python
df_produtos_unicos['produto_id'] = df_produtos_unicos.index + 1
df_produtos_unicos.head()

# Incluindo coluna de id de cada produto único
```

```python
df_mp_original = df_mp_original.merge(df_produtos_unicos, on=['produto'], how='left')
df_mp_original.head()

# Incluindo coluna de id de cada produto
```

```python
with open ('insert_meses.sql', 'w', encoding='utf-8') as arquivo:
    for _, row in df_mes.iterrows():
        arquivo.write(f"INSERT INTO meses (mes) VALUES ('{row['mes']}');\n")
```

```python
with open ('insert_anos.sql', 'w', encoding='utf-8') as arquivo:
    for _, row in df_ano.iterrows():
        arquivo.write(f"INSERT INTO anos (ano) VALUES ('{row['ano']}');\n")
```

```python
with open ('insert_regioes.sql', 'w', encoding='utf-8') as arquivo:
    for _, row in df_regioes_unicas.iterrows():
      regiao = row['regiao'].replace("'", "''")
      arquivo.write(f"INSERT INTO regioes (regiao) VALUES ('{row['regiao']}');\n")
```

```python
with open ('insert_estados.sql', 'w', encoding='utf-8') as arquivo:
    for _, row in df_estados_unicos.iterrows():
      estado = row['estado'].replace("'", "''")
      arquivo.write(f"INSERT INTO estados (estado) values ('{row['estado']}');\n")
```

```python
with open ('insert_produtos.sql', 'w', encoding='utf-8') as arquivo:
    for _, row in df_produtos_unicos.iterrows():
        produto = row['produto'].replace("'", "''")
        arquivo.write(f"INSERT INTO produtos (produto) values ('{row['produto']}');\n")
```

```python
with open('insert_biocombustiveis.sql', 'w', encoding='utf-8') as arquivo:
    for _, row in df_mp_original.iterrows():
      quantidade_m3 = int(row['quantidade_m3'])
      ano_id = int(row['ano_id'])
      mes_id = int(row['mes_id'])
      regiao_id = int(row['regiao_id'])
      estado_id = int(row['estado_id'])
      produto_id = int(row['produto_id'])

      arquivo.write(
          f"INSERT INTO biocombustiveis (quantidade_m3, ano_id, mes_id, regiao_id, estado_id, produto_id)"
          f"VALUES ('{quantidade_m3}', '{ano_id}', '{mes_id}', '{regiao_id}', '{estado_id}', '{produto_id}');\n"
      )
```
